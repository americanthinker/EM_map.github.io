```python
from collections import Counter
```

```python
from data_processing import punc_removal as pr
```

```python
joined = ' '.join(locations.tolist())
```

```python
len(joined)
```

```python
cleaned = pr(joined)
len(cleaned)
```

```python
places = Counter(cleaned.split())
```

```python
garbage_words = ['or', 'to', 'the', 'am', 'East', 'West', 'Coast', 'will', 'my', 'flexible', 'can', 'most', 'prefer', 'choice']
```

```python
for word in garbage_words:
    if word in places:
        del places[word]
```

```python
len(places)
```

```python
dfplaces = pd.DataFrame(places, index=range(1))
```
